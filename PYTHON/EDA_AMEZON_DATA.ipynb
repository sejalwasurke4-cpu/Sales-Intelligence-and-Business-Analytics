```python
# Find out Total Revenue as per Category .
```

```python
df = pd.merge(orders_df,
              products_df,
              on = 'P_ID',
              how = 'left')
df
              
     
```

```python
df['Revenue'] = df['Qty']*df['Price']
df['Revenue']
```

```python
cetegory_revenue = df.groupby('Category')['Revenue'].sum().sort_values(ascending = False)
cetegory_revenue
```

```python
plt.bar(cetegory_revenue.index , cetegory_revenue.values)
plt.title("Category wise Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
# Conclusion - This chart show balance performence across a category wise 
#revenue with no standout high and low performer.
```

```python
# -- Which products have the highest order volume
```

```python
Product_volume = df.groupby("Category")["Qty"].sum()
Product_volume 
```

```python
plt.pie(Product_volume.values,labels=Product_volume.index , autopct="%0.0f%%")
plt.title("Category wise Volume")
plt.savefig('bar1.jpeg')
plt.show()
```

##### Insight - This analysis found that allmost categories are balanced in market .

```python
#Order volume vs revenue growth
```

```python
df['Year'] = df['Order_Date'].dt.year
df['Year'] 
```

```python
order_revenue = df.groupby(['Or_ID','Year'])['Revenue'].sum().reset_index()
order_revenue                    
```

```python
aov = order_revenue.groupby('Year')['Revenue'].mean().reset_index(name='AOV')

aov
```

```python
plt.plot(aov['Year'] ,aov['AOV'] , marker = 'o')
```

```python
#- This AOV shows indicate customer are buying more expencive products or placing biggr orders. 
```
